## Calculating cost with gradient descent and learning rate
- Change the iteration and learning rate vaules and see the impact on cost. 
- Low iteration values with high learning rate (i.e. big steps) may lead to miss the global minimum
- Goal is to reach minimum cost with minimum iteration

```python
# code credit:codebasics https://codebasics.io/coming-soon

import numpy as np

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 100       #change value
    n = len(x)
    learning_rate = 0.08   #change value

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)


#m_curr = b_curr = 0  # Initial slope and intercept
#iterations = 100  # Total updates
#learning_rate = 0.08  # Controls update size per iteration
#
# Gradient Descent Loop
#y_predicted = m_curr * x + b_curr  # Prediction using current line
#cost = (1/n) * sum((y - y_predicted)**2)  # Mean squared error (MSE)
#md = -(2/n)*sum(x*(y-y_predicted))  # Gradient w.r.t slope
#bd = -(2/n)*sum(y-y_predicted)  # Gradient w.r.t intercept

# Parameters update rule
#m_curr = m_curr - learning_rate * md
#b_curr = b_curr - learning_rate * bd

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Sample Dataset
x = np.array([1, 2, 3, 4, 5], dtype=np.float64)
y = np.array([5, 7, 9, 11, 13], dtype=np.float64)

# Step 2: Hyperparameters
iterations = 1000          # Number of training steps
learning_rate = 0.028       # Size of the step to update parameters

# Step 3: Initialization
m = 0.02                      # Initial slope
b = 0.05                      # Initial intercept

n = len(x)
cost_history = []
m_history = []
b_history = []

# Step 4: Gradient Descent Loop
for i in range(iterations):
    y_pred = m * x + b
    cost = (1/n) * np.sum((y - y_pred) ** 2)  # Mean Squared Error (MSE)
    
    # Partial derivatives
    md = -(2/n) * np.sum(x * (y - y_pred))  # Gradient for slope
    bd = -(2/n) * np.sum(y - y_pred)        # Gradient for intercept

    # Update parameters
    m = m - learning_rate * md
    b = b - learning_rate * bd

    # Store for visualization
    cost_history.append(cost)
    m_history.append(m)
    b_history.append(b)

    print(f"Iteration {i+1}: m={m:.4f}, b={b:.4f}, cost={cost:.4f}")

# Step 5: Visualization - Regression Line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', label='Actual Data')
plt.plot(x, m * x + b, color='blue', label=f'Fitted Line: y = {m:.2f}x + {b:.2f}')
plt.title('Linear Regression Result')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Visualization - Cost over Iterations
plt.figure(figsize=(10, 5))
plt.plot(range(1, iterations + 1), cost_history, color='green')
plt.title('Cost Reduction Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error (Cost)')
plt.grid(True)
plt.show()

# Step 7: Visualization - Slope (m) and Intercept (b) convergence
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(m_history, label='Slope (m)')
plt.title('Slope Convergence')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(b_history, label='Intercept (b)', color='orange')
plt.title('Intercept Convergence')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.grid(True)

plt.tight_layout()
plt.show()

# Step 8: Final Output
print(f"\nFinal model: y = {m:.4f}x + {b:.4f}")
```
